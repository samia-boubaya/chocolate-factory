````python
# Initialize dataframe totals
totals = pd.DataFrame()

# Calculate total sales by product name
totals["sales"] = df.groupby('Product Name')['Sales'].sum()
print("\nTotal Sales by Product Name -------------------------")
print(totals["sales"])

# Calculate total cost by product name
totals["cost"] = df.groupby('Product Name')['Cost'].sum()
print("\nTotal Cost by Product Name -------------------------")
print(totals["cost"])

# Calculate total gross profit by product name
totals["gross_profit"] = df.groupby('Product Name')['Gross Profit'].sum()
print("\nTotal Gross Profit by Product Name -------------------------")
print(totals["gross_profit"])

# Calculate total units by product name
totals["units"] = df.groupby('Product Name')['Units'].sum()
print("\nTotal Units by Product Name -------------------------")
print(totals["units"])

# Make "Product Name" a column instead of the index
totals = totals.reset_index()  # <-- this is the key step

# Optional: sort by gross profit descending
totals = totals.sort_values(by="gross_profit", ascending=False)

# Save to CSV Product Totals
totals.to_csv("../data/interim/product_totals.csv", index=False)

# View the dataframe
totals.head(totals.shape[0])
````

````python

import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load CSV with Factory info
df = pd.read_csv("../data/raw/wonka_choc_factory.csv") 

# Initialize geocoder
geolocator = Nominatim(user_agent="chocolate_factory_app")
geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)  # avoid rate limits

# Function to get address components
def get_factory_address(lat, lon):
    try:
        location = geocode((lat, lon), exactly_one=True)
        if location is None:
            return pd.Series([None, None, None, None])
        address = location.raw.get('address', {})
        country = address.get('country')
        region = address.get('state') or address.get('region')
        city = address.get('city') or address.get('town') or address.get('village')
        postcode = address.get('postcode')
        return pd.Series([country, region, city, postcode])
    except:
        return pd.Series([None, None, None, None])

# Apply function to all rows
df[['Factory Country', 'Factory Region', 'Factory City', 'Factory Postal Code']] = df.apply(
    lambda row: get_factory_address(row['Latitude'], row['Longitude']), axis=1
)

# Save updated CSV
df.to_csv("../data/interim/factory_address.csv", index=False)

df.head()
````
